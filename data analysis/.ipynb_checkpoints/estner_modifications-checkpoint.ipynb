```python
def create_phrase(words):
    '''Takes a list of words and creates a phrase where only the last word is lemmatized'''
    phrase = [word.lower() for word in words[:-1]]
    phrase.append(Text(words[-1]).lemmas[0].split('|')[0].lower())
    return ' '.join(phrase)
```

```python
def create_lemmatized_phrase(words):
    '''Takes a list of words and creates a phrase where all words are lemmatized. Meant for
    compatibility with estner gazetteer lookup.'''
    phrase = ' '.join(words)
    lemmatized_phrase = [lemmas.split('|')[0].lower() for lemmas in Text(phrase).lemmas]
    return ' '.join(lemmatized_phrase)
```

```python
vocation_label_b = 'B-VOC'
vocation_label_i = 'I-VOC'
modified_counter = 0
look_ahead = 3
overlap_with_titles = Counter()

with open('estner.json') as estnltk_json:
    with open('estner_custom.json', 'w') as custom_estnltk_json:
        line_count = 0
        for line in estnltk_json:
            estnltk_dict = json.loads(line.strip())
            words = estnltk_dict.get('words', [])
            for i in range(len(words)):
                # Need to make sure we are skipping over already found phrases
                # in order not to detect subsequent subphrases as beginnings of a new entity
                longest_found_phrase = 1
                if i < len(words)-1:
                    max_look_ahead = (look_ahead if len(words)-1-i >= look_ahead else len(words)-1-i)
                    for j in range(i+1, i+1+max_look_ahead):
                        list_of_words = [word['text'] for word in words[i:j]]
                        phrase = create_phrase(list_of_words)
                        #print('Looking for:', phrase)
                        
                        if phrase in verified_vocations:
                            #print('FOUND: ', phrase)
                            longest_found_phrase = len(list_of_words)
                            if phrase in verified_titles:
                                overlap_with_titles[phrase] += 1
                            words[i]['label'] = vocation_label_b
                            words[i]['validated'] = False
                            modified_counter += 1
                            for w in words[i+1:j]:
                                w['label'] = vocation_label_i
                                w['validated'] = False
                                modified_counter += 1
                else:
                    phrase = create_phrase([words[i]['text']])
                    #print('Looking for\t', phrase)    
                    if phrase in verified_vocations:
                        #print('FOUND: ', phrase)
                        if phrase in verified_titles:
                            overlap_with_titles[phrase] += 1      
                        words[i]['label'] = vocation_label_b
                        words[i]['validated'] = False
                        modified_counter += 1
                i += longest_found_phrase
            line_count += 1
            #if line_count == 3:
            #   break

            json.dump(estnltk_dict, custom_estnltk_json)
            custom_estnltk_json.write('\n')
```

```python
# Analyze existing lexicons
existing_gazetteer = {'per' : set(), 'org' : set(), 'loc' : set(), 
                      'peop' : set(), 'fac' : set()}
edge_cases = []

with open('gazetteer.txt', 'r', encoding='utf-8') as f:
    for line in f:
        entity, tag = line.split('\n')[0].split('\t', maxsplit=1)
        #print(entity, tag)
        if tag in existing_gazetteer:
            existing_gazetteer[tag].add(entity)
        else:
            edge_cases.append((entity, tag))
```

```python
i = 0
cleaned_locs = set()
for entity in existing_locs:
    if 'saard' in entity:
        cleaned_locs.add(re.sub('saard', 'saared', entity))
    else:
        cleaned_locs.add(entity.strip('.'))
    i += 1
```

```python
pattern = re.compile('\s?\d\S*\s?')
to_be_removed = set()
to_be_added = set()
for entity in cleaned_locs:
    if re.search(pattern, entity):
        to_be_removed.add(entity)
        to_be_added.add(re.sub(pattern, '', entity).strip())
```

```python
cleaned_peop = set()
removed_parts = []
# If this pattern gets matched, remove the entry from gazetteer
removal_pattern = re.compile('^list\s|:|;|!|\sof\s|\sand\s|\sfor\s')
# If this pattern gets matched, replace the match with '' but keep
# the entry
replacement_pattern = re.compile('\(.*\)')
for entity in existing_peop:
    parts = entity.split(',', maxsplit=1)
    for part in parts:
        if re.search(removal_pattern, part):
            removed_parts.append(part)
        else:
            cleaned_peop.add(re.sub(replacement_pattern, '', part).strip()) 
```

```python
# Create gazetteer with cleaned versions of existing entities
# Temporarily substitute 'peop' tags with 'per' until we start
# actually using 'peop' to distinguish public personas
with open('gazetteer_cleaned.txt', 'w', encoding='utf-8') as f:
    for name in cleaned_peop_last:
        f.write(name.strip())
        f.write('\tper\n')
    for name in existing_pers:
        f.write(name.strip())
        f.write('\tper\n')
    for place in cleaned_locs:
        f.write(place.strip())
        f.write('\tloc\n')
    for org in cleaned_orgs:
        f.write(org.strip())
        f.write('\torg\n')
    for title in verified_titles:
        f.write(title.strip())
        f.write('\ttitle\n')
    for vocation in verified_vocations:
        f.write(vocation.strip())
        f.write('\tvoc\n')
```

```python
#title_label = 'TITLE'
#total_counts = Counter()
label_counts = Counter()
total_words = 0
doc_count = 0
doc_lengths = []
punctuation_count = 0
alt_total_count = 0

with open('data/estner.json') as estnltk_json:
    #with open('estner_custom2.json', 'w') as custom_estnltk_json:
    i = 0
    for line in estnltk_json:
        estnltk_dict = json.loads(line.strip())
        doc_count += 1
        doc_length = 0
        for word in estnltk_dict.get('words', []):
            total_words += 1
            label_counts[word['label']] += 1
            doc_length += 1
        doc_lengths.append(doc_length)
        
        text = Text(estnltk_dict.get('text', []))
        text.tag_analysis()
        for word in text['words']:
            for analysis in word['analysis']:
                if analysis['partofspeech'] == 'Z':
                    punctuation_count += 1
                    break
            alt_total_count += 1
        #break
        #if i % 10:
        #    print(estnltk_dict['text'])
        i += 1
```

```python
# Split the corpus into train and test sets
with open('estner.json') as estnltk_json:
    with open('estner_train.json', 'w') as train:
        with open('estner_test.json', 'w') as test:
            i = 0
            train_count = 0
            test_count = 0
            for line in estnltk_json:
                estnltk_dict = json.loads(line.strip())
                if i % 10 == 0:
                    json.dump(estnltk_dict, test)
                    test.write('\n')
                    test_count += 1
                else:
                    json.dump(estnltk_dict, train)
                    train.write('\n')
                    train_count += 1
                i += 1
```

```python
test_label_counts = Counter()
test_total_words = 0
test_labels = []

with open('estner_test.json') as test_json:
    with open('estner_test_no_labels.json', 'w') as no_labels:
        for line in test_json:
            estnltk_dict = json.loads(line.strip())
            doc_labels = []
            new_words = []
            for word in estnltk_dict.get('words', []):
                test_total_words += 1
                test_label_counts[word['label']] += 1
                doc_labels.append(word['label'])
                new_word = word
                del new_word['label']
                new_words.append(new_word)

            test_labels.append(doc_labels) 
            estnltk_dict['words'] = new_words
            json.dump(estnltk_dict, no_labels)
            no_labels.write('\n')
```
